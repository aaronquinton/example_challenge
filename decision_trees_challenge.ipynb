# Introduction to Decision Trees - Challenge

```python
# Load modules
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
# Input training data for decision tree
X = pd.DataFrame({'Commute_in_Car':[0,1,0,0,1,1],
                  'Wake_up_Time(hrs)':[6.75,8,7.25,7.75,8,8.5]})
y = [0,0,1,1,0,1]

labels = ['On Time', "Late"]
```

```python
# View Input Data
X
```

```python
# View Labels
y
```

```python
# Train Decision tree model
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
# Visualize Decision Tree
import graphviz

dot_data = export_graphviz(model, feature_names=X.columns, class_names=labels)
graphviz.Source(dot_data)
```
